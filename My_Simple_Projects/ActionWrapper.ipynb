Task:
Intervene in the stream of actions sent by the agent and, with a probability of 80%, replace the current action with random one. By issuing the random actions, we make our agent explore the environment and from time to time drift away from the beaten track of its policy

```python
import gym
import random
```

```python
class RandomActionWrapper(gym.ActionWrapper):
    
    def __init__(self, env, epsilon=0.8):
        
        super(RandomActionWrapper, self).__init__(env)
        self.epsilon = epsilon
    
    def action(self, action):
        if random.random() < self.epsilon:
            print("Random!")
        
            return self.env.action_space.sample()
        
        return action
```

```python
if __name__ == "__main__":
    env = RandomActionWrapper(gym.make("CartPole-v0"))
    obs = env.reset()
    total_reward = 0.0
    
    while True:
        obs, reward, done, _ = env.step(0)
        total_reward += reward 
        
        if done:
            break
    
    print("Reward got: %.2f" % total_reward)
```

15/17 almost 80% of time we get the random selection of action
